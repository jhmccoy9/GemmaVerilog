```python
import os
import random

def test(dirpath, generator, k: int):
    # generate list of all file paths in the directory path that include "prompt" in the name
    prompt_files = []
    for root, dirs, files in os.walk(dirpath):
        for file in files:
            if "prompt" in file:
                prompt_files.append(os.path.join(root, file))

    if not prompt_files:
        print(f"No files containing 'prompt' found in {dirpath}")
        return



    # MUST BE PROMPT 3
    random_file_path = prompt_files[0]
    for i in prompt_files:
      if "3" in i:
        random_file_path = i

    # read the prompt text
    try:
        with open(random_file_path, 'r') as f:
            prompt_text = f.read()
    except IOError as e:
        print(f"Error reading file {random_file_path}: {e}")
        return

    # Extract just the filename without extension (e.g. 'prompt1_mux')
    # os.path.basename gets the file name from the full path
    # os.path.splitext removes the .v extension
    filename_no_ext = os.path.splitext(os.path.basename(random_file_path))[0]

    for i in range(1, k + 1):
        # Generate outputs
        generated_content = generator([{"role": "user", "content": prompt_text}], max_new_tokens=2048, return_full_text=False)[0]["generated_text"]

        # Get model name and replace slashes with underscores (e.g. "meta-llama/Llama-2" -> "meta-llama_Llama-2")
        # so it doesn't create invalid directory paths
        model_name = generator.model.name_or_path.replace("/", "_")

        # Construct the new path pointing strictly to /content/
        output_filename = f"{filename_no_ext}-{model_name}-{i}.v"
        output_file_path = os.path.join("/content", output_filename)

        try:
            with open(output_file_path, "w") as f:
                f.write(generated_content)
            print(f"Saved output to: {output_file_path}")
        except IOError as e:
            print(f"Error writing file {output_file_path}: {e}")
```
